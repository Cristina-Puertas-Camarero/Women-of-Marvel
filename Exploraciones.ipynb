```python
import pandas as pd
from odf.opendocument import load
from odf.table import Table, TableRow, TableCell
```

```python
def read_ods(file_path):
    doc = load(file_path)
    sheet = doc.spreadsheet.getElementsByType(Table)[0]

    data = []
    for row in sheet.getElementsByType(TableRow):
        cells = [cell.getElementsByType(TableCell)[0].textContent for cell in row.getElementsByType(TableCell)]
        data.append(cells)

    return pd.DataFrame(data)
```

```python
%pip install pyexcel-ods3

import pyexcel_ods3

# Leer el archivo ODS
data = pyexcel_ods3.get_data('marvel_superheroes.ods')

# Verificar los nombres de las hojas
print(data.keys())

# Convertir la hoja de cálculo en un DataFrame de pandas
sheet_name = 'Hoja1'  # Cambia esto al nombre de tu hoja si es diferente
df_los30mejoresmixto = pd.DataFrame(data[sheet_name])

# Mostrar el DataFrame
print(df_los30mejoresmixto)
```
