```python
#vamos a limpiar cada csv, para que sea más fácil trabajar con ellos
#En la columna nombre  vamos a homogeneizar los nombres
#Vamos a poner en la columna name, todos los nombres en minusclas. Los espacios los vamos a sustitir por guion bajo.
#Vamos a quitar los acentos y las tildes
```

```python
import pandas as pd

          
```

```python
#cargar csv
df = pd.read_csv('mixto_30_popularidad_ventas.csv')
#lo guardamos con el nombre df_quienvendemas
df_quienvendemas = df
#vamos a establecer como columnas la primera fila
df_quienvendemas.columns = df_quienvendemas.iloc[0]
```

```python
#vamos a limpiar las columnas Woman y Men-M
df_quienvendemas[['Woman', 'Men-M']] = df_quienvendemas[['Woman', 'Men-M']].apply(lambda x: x.str.lower())
df_quienvendemas[['Woman', 'Men-M']] = df_quienvendemas[['Woman', 'Men-M']].apply(lambda x: x.str.replace(' ', '_'))
#vamos a quitar los acentos y las tildes
df_quienvendemas[['Woman', 'Men-M']] = df_quienvendemas[['Woman', 'Men-M']].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
#guardamos el csv
df_quienvendemas.to_csv('quien_vende_mas.csv', index=False)
```

```python
#cargar siguiente csv   
df = pd.read_csv('nombre_poder_total.csv')  
#lo guardamos con el nombre df_marvelcaracteristicas
df_marvelcaracteristicas = df
#vamos a limpiar la columna name    
df_marvelcaracteristicas['Name'] = df_marvelcaracteristicas['Name'].str.lower()
df_marvelcaracteristicas['Name'] = df_marvelcaracteristicas['Name'].str.replace(' ', '_')
#vamos a quitar los acentos y las tildes
df_marvelcaracteristicas['Name'] = df_marvelcaracteristicas['Name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
#guardamos el csv
df_marvelcaracteristicas.to_csv('marvel_caracteristicas.csv', index=False)
```

```python
#cargar siguiente csv
df = pd.read_csv('Id_nombre_genero.csv')
#lo guardamos con el nombre df_nombres_genero
df_nombres_genero = df      
#vamos a limpiar la columna Name    
df_nombres_genero['Name'] = df_nombres_genero['Name'].str.lower()
df_nombres_genero['Name'] = df_nombres_genero['Name'].str.replace(' ', '_')
#vamos a quitar los acentos y las tildes
df_nombres_genero['Name'] = df_nombres_genero['Name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
#guardamos el csv
df_nombres_genero.to_csv('nombres_genero.csv', index=False)
```
